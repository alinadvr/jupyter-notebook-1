## Пам'ятки Гайани

### Місто Джорджтаун
<p style="border: solid 1px blue; padding: 20px;">Це <span style="color: red; font-weight: bold;">столиця</span> Гайани, її торговельний і політичний центр. Місто розташоване на березі Атлантичного океану. Круглий рік в ньому тепле і вологе повітря. Джорджтаун заснували в 1781 році голландські колоністи, а потім його захопили британці. У місті багато індуїстських храмів, дерев’яних будівель, є собор, маяк, музеї.</p>

### Водоспад Кайетур
<p style="border: solid 1px blue; padding: 20px;">Водоспад розташований на заході Гайани на річці Потаро. Він є одним з найбільших водоспадів у світі. Його висота становить 226 метрів. Ніагарський водоспад нижче майже в 5 разів, а водоспад Вікторія в 2 рази. Кайетур не дуже популярний через свого важкодоступного місця розташування. Він оточений тропічним лісом, навколо нього збереглася первозданна природа.</p>
